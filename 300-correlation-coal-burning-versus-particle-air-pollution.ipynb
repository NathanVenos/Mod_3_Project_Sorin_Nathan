## Merge pollution data with electricity from fossil/coal data

The data will be aggregated for the entire 2010-2015 period because it does not change significantly for any given country. This increases the sample size for the statistical tests below.

Scatter plots of the PM25 air pollution versus energy from fossil/coal sources are shown to illustrate the respective correlations.

```python
pollution_melt_df = pollution_df.melt(
    id_vars=["Country Name", "Country Code", "Region", "IncomeGroup"],
    var_name="Year",
    value_name="PM25"
)

fossil_melt_df = fossil_df.melt(
    id_vars=["Country Name", "Country Code", "Region", "IncomeGroup"],
    var_name="Year",
    value_name="fossil_frac"
)

coal_melt_df = coal_df.melt(
    id_vars=["Country Name", "Country Code", "Region", "IncomeGroup"],
    var_name="Year",
    value_name="coal_frac"
)

df = pd.merge(pollution_melt_df, fossil_melt_df, how="inner")
df = pd.merge(df, coal_melt_df, how="inner")

df
```

```python
f, ax = plt.subplots(figsize=(12, 8))

ax = sns.scatterplot(data=df, y="fossil_frac", x="PM25", hue="Region",
                   ax=ax, s=60, alpha=0.6)#, palette="cividis");

ax.axvline(x=35, ymin=0, ymax=1, color='red');
ax.annotate(s="long-term health effects", xy=(34, 5), xytext=(20, 20),
            textcoords='offset points', color="red", fontsize=16);

ax.set_title("Correlation PM25-air-pollution versus fossil-burning (2010-2015)", fontsize=20);
ax.set_xlabel("population-weighted, mean annual exposure: PM25 concentration (ug/m3)", fontsize=16);
ax.set_ylabel("electricity production from fossil sources (% of total)", fontsize=16);
ax.tick_params(axis='both', which='major', labelsize=14)
ax.legend(fontsize=12);
ax.set_xlim(-5, 120);

plt.savefig('500-correlation-pm25-fossil.png', dpi=300, bbox_inches='tight')
```

## Statistical analysis of PM25-fossil/coal correlations

**METHODOLOGY**

 * Countries are analyzed for the aggregated period 2010-2015 and grouped by geographical region.
 * The observed Pearson correlation coefficient is calculated for region.
 * The Bootstrap resampling method is then employed within each region to estimate the 95% confidence interval for the correlation coefficient.
 * The Randomization resampling method is then utilized to find if a *positive correlation* exists. The exact hypothesis formulation is shown below.

**HYPOTHESIS**

 * $H_0$ assumes that no correlation exists ($R = 0$)
 * $H_a$ assumes that a positive correlation exists ($R > 0$)
 * $\alpha$ is set to 0.05
 
**INTERPRETATION**

 * If the null hypothesis is *not rejected*, then other factors *must* contribute to the PM25 air pollution, such as natural sources (e.g. deserts and active volcanoes), the lack of environmental laws, or corruption.
 * If the null hypothesis is *rejected*, then industrial activity most likely explains PM25 air pollution.
 
**FUTURE STUDIES**

Additional features must be considered here in order to explain the different sources contributing to the PM25 air pollution. The next features to be investigated are:

 * The quantities of fossil/coal fuel consumed by each country.
 * The status of environmental laws (especially related to air quality) in each country.
 * The corruption levels in each country.

```python
# PM25 versus fossil consumption

for index in [0, 3, 4, 5, 6, 7]:
    
    # identify region
    region = df.Region.value_counts(dropna=False).index[index]
    print()
    print(region)
    
    # filter by region
    df_temp = df[df.Region == region]
    
    # select features
    pollution = df_temp.PM25
    energy = df_temp.fossil_frac

    corr = sorin.calculate_correlation(zip(pollution, energy))
    print(F"Pearson correlation = {corr:.3f}")

    conf_int = sorin.resample_correlation_bootstrap(zip(pollution, energy), iterations=10000)
    print(F"95% CI = ({conf_int[0]:.3f}, {conf_int[1]:.3f})")

    pvalue = sorin.resample_correlation_randomize(zip(pollution, energy), iterations=10000)
    print(F"p-value = {pvalue:.3f}")
```

```python
# PM25 versus coal consumption

for index in [0, 3, 4, 5, 6, 7]:
    
    # identify region
    region = df.Region.value_counts(dropna=False).index[index]
    print()
    print(region)
    
    # filter by region
    df_temp = df[df.Region == region]
    
    # select features
    pollution = df_temp.PM25
    energy = df_temp.coal_frac

    corr = sorin.calculate_correlation(zip(pollution, energy))
    print(F"Pearson correlation = {corr:.3f}")

    conf_int = sorin.resample_correlation_bootstrap(zip(pollution, energy), iterations=10000)
    print(F"95% CI = ({conf_int[0]:.3f}, {conf_int[1]:.3f})")

    pvalue = sorin.resample_correlation_randomize(zip(pollution, energy), iterations=10000)
    print(F"p-value = {pvalue:.3f}")
```

## About the bootstrap/randomized resampling methods<br> to study the correlation coefficients

```python
xs = np.arange(11) + np.random.normal(loc=0, scale=1, size=11)
ys = np.arange(11) + np.random.normal(loc=0, scale=1, size=11)
# ys = ["a","a","a","a","a","a","a","a","a","a","a"]
data = list(zip(xs, ys))

print(F"correlation = {sorin.calculate_correlation(data)}")

resample = sorin.resample_correlation_bootstrap(data=data, iterations=10000)
print(F"95% CI = {resample})")

resample = sorin.resample_correlation_randomize(data=data, iterations=10000)
print(F"p-value = {resample}")
```
